```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("CashFlow.csv", header=None, index_col=0)
df.head()
```

```python
NI = df.loc['Net Income'].tolist()
NI

# Chuyển các chuỗi thành số nguyên và chia để lấy đơn vị triệu
converted_numbers_NI = [int(num) / 1000000000 for num in NI]

# In ra kết quả
print(converted_numbers_NI)
```

```python
OA = df.loc['Net Cash Provided By Operating Activities'].tolist()
OA

# Chuyển các chuỗi thành số nguyên và chia để lấy đơn vị triệu
converted_numbers_OA = [int(num) / 1000000000 for num in OA]

# In ra kết quả
print(converted_numbers_OA)
"""Net Cash Provided By Operating Activities (Tiền mặt thuần từ hoạt động kinh doanh) 
là số tiền mặt mà một doanh nghiệp tạo ra từ các hoạt động kinh doanh chính, 
như bán hàng hoặc cung cấp dịch vụ, sau khi trừ đi các chi phí hoạt động như trả lương, 
mua nguyên liệu, và thanh toán các chi phí khác.

Nói đơn giản, đây là số tiền thực tế mà doanh nghiệp thu về từ các hoạt động kinh doanh 
hàng ngày sau khi đã trả hết các chi phí liên quan trực tiếp đến việc kinh doanh."""
```

```python
FA = df.loc['Net Cash Used Provided By Financing Activities'].tolist()
FA 
# Chuyển các chuỗi thành số nguyên và chia để lấy đơn vị triệu
converted_numbers_FA = [int(num) / 1000000000 for num in FA]
# In ra kết quả
print(converted_numbers_FA)
```

```python
IA = df.loc['Net Cash Used For Investing Activites'].tolist()
IA
# Chuyển các chuỗi thành số nguyên và chia để lấy đơn vị triệu
converted_numbers_IA = [int(num) / 1000000000 for num in IA]
# In ra kết quả
print(converted_numbers_IA)
```

```python
# Lấy hàng 'date'
date_list = pd.to_datetime(df.loc['Date'].tolist())
date_list
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(date_list, converted_numbers_NI, marker='o', label='Thu nhập ròng (Net Income)', color='green')
plt.plot(date_list, converted_numbers_OA, marker='o', label='Tiền mặt ròng được cung cấp bởi các hoạt động kinh doanh (Operating)', color='blue', linestyle = 'dashdot')
plt.plot(date_list, converted_numbers_FA, marker='o', label='Tiền mặt ròng dùng cho các hoạt động tài chính (Financing)', color='yellow', linestyle = 'dashdot')
plt.plot(date_list, converted_numbers_IA, marker='o', label='Tiền mặt ròng dùng cho hoạt động đầu tư (Investing)', color='red', linestyle = 'dashdot')

# Thêm tiêu đề và nhãn
#plt.title('Net Income', fontsize=14)
plt.xlabel('năm', fontsize=12)
plt.ylabel('tỉ đô $', fontsize=12)
plt.xticks(rotation=45)
plt.grid()
# Định dạng trục y với đơn vị triệu
#plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))
# Function add a legend
plt.legend()
# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

# các hoạt động Phân bổ phần trăm Tiền mặt ròng từ hoạt động tài chính năm 2023 là gì ?

```python
import matplotlib.pyplot as plt

# Dữ liệu cho biểu đồ
labels = ['Trả nợ', 'Mua lại cổ phiếu', 'Chi trả cổ tức', 'Hoạt động tài chính khác']
values = [-5_923_000_000,  -82_981_000_000, -15_025_000_000, -4_559_000_000]

# Tính phần trăm
total = sum([v for v in values])
percentages = [v / total * 100 for v in values]

# Vẽ biểu đồ tròn
plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=90, 
        colors=['red','blue','yellow','green'], pctdistance=0.85, 
        explode=(0.05, 0.05, 0.05, 0.05))

# Tạo biểu đồ hình tròn và tiêu đề
plt.gca().set_aspect('equal')  # Đảm bảo biểu đồ là hình tròn
plt.title("Phân bổ phần trăm Tiền mặt ròng từ hoạt động tài chính năm 2023")
plt.show()
```

- Doanh nghiệp tập trung mạnh vào việc mua lại cổ phiếu, điều này có thể là tín hiệu tích cực về niềm tin của ban lãnh đạo vào giá trị cổ phiếu.

- Chi trả cổ tức cũng khá cao, đảm bảo lợi ích của cổ đông.

- Trả nợ và các hoạt động khác ít được chú trọng trong năm tài chính này
