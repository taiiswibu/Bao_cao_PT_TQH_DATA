```python
import pandas as pd

# Đọc dữ liệu từ tệp Excel
file_path = r'D:\Code\Bao_cao_PT_TQH_DATA\data\AAPL.xlsx'  # Đường dẫn tới tệp Excel
df = pd.read_excel(file_path, sheet_name='AAPL - Income Statement FY')  # Đọc trang cụ thể

# Lưu dữ liệu thành tệp CSV
csv_file_path = r'D:\Code\Bao_cao_PT_TQH_DATA\data\AAPL_income_statement.csv'  # Đường dẫn tới tệp CSV cần lưu
df.to_csv(csv_file_path, index=False, encoding='utf-8-sig')  # Lưu tệp CSV mà không có chỉ số

print(f"Dữ liệu đã được lưu vào {csv_file_path}")

```

```python
data = pd.read_csv(csv_file_path)
null_count = data.isnull().values.sum()  # Tính tổng số giá trị null trong DataFrame
print(f"Số lượng giá trị null trong DataFrame: {null_count}")
```

```python
data = data.dropna()  # Xóa các hàng có giá trị null
data.head()
```

```python
row= data.iloc[0]  # Lấy hàng có chỉ số 3 (hàng thứ 2)
row
```

```python
# Đổi tên các cột trong DataFrame
data.rename(columns=row, inplace=True)
data.head()
```

```python
# Xóa hàng với chỉ số là 3
data.drop(index=3, inplace=True)
# Hiển thị DataFrame 
data.head()
```
