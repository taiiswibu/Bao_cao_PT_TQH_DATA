```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = r'D:\Code\Bao_cao_PT_TQH_DATA\data\AAPL - Key Metrics FY.xlsx'
df = pd.read_excel(file, header=None, index_col=0)
df.head()
```

```python
PT = df.loc["Payables Turnover"].values
# thu gọn dữ liệu
converted_PT =np.round(PT.astype(float), 2)
converted_PT
```

```python
date = pd.to_datetime(df.loc['Date'].tolist())
date
```

```python
#vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(date, converted_PT, marker='o', color='red')
plt.xlabel('Năm', fontsize=12)
plt.ylabel('Vòng', fontsize=12)
plt.title('Vòng quay các khoản phải thu', fontsize=14)
plt.grid()
plt.legend()
plt.show()
```

Vòng quay tài khoản phải thu = Doanh số tín dụng ròng / Các khoản phải thu bình quân

Trong đó:

Doanh số tín dụng ròng =  Doanh số bán hàng tín dụng – Doanh thu bán hàng trả lại – Phụ cấp bán hàng

Các khoản phải thu bình quân =(các khoản phải thu bắt đầu + các khoản phải thu kết thúc)/2

```python
IT = df.loc["Inventory Turnover"].values
# thu gọn dữ liệu
converted_IT =np.round(IT.astype(float), 2)
converted_IT
```

```python
plt.figure(figsize=(10, 6))
plt.plot(date, converted_IT, marker='o', color='blue', linestyle = 'dashdot')
plt.xlabel('Năm', fontsize=12)
plt.ylabel('Vòng', fontsize=12)
plt.title('Vòng quay hàng tồn kho', fontsize=14)
plt.grid()
plt.legend()
plt.show()
```

Vòng quay hàng tồn kho = Doanh số bán hàng / Các khoản phải thu bình quân
